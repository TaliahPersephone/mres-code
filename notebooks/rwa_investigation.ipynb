```python
def gaus_prob(x, mu, sigma):
    return 1/(sigma * np.sqrt(2*np.pi)) * np.exp(-np.pow((x - mu),2)/ (2 * sigma ** 2))

def f(x):
    return 1/ (1 + x)

mu = 1
sigma = 1
samples = np.int64(1e4)
span = 5

sum(gaus_prob(x, mu, sigma**2) * f(x) for x in np.linspace(mu-span, mu+span, samples)) * 2 * span / samples
```
